```python
# Step (1)
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
# Step (2)
iris = sns.load_dataset('iris')
iris.head()
```

```python
# Step (3)
iris = iris.drop('species', axis=1)
iris.head()
```

```python
# Step (4)
iris['sepal_sum'] = iris['sepal_length'] + iris['sepal_width']
iris['petal_sum'] = iris['petal_length'] + iris['petal_width']
iris.head()
```

```python
# Step (5)
summary_stats = pd.DataFrame({
    'mean': iris.mean(),
    'std': iris.std(),
    'min': iris.min(),
    'max': iris.max()})
```

```python
# Display the summary stats DataFrame
print(summary_stats)
```
